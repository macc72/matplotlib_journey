### A basic exampe where all parameters are used for the line graph

```python
# Sample data
x = [0, 1, 2, 3, 4, 5]
y = [0, 1, 4, 9, 16, 25]

# Plotting the line graph
plt.plot(x, y,  
         # Line paramters
         color='green', # line color
         linestyle='-', # Line style: solid (-), dashed (--), dashdot (-.), dotted (:)        
         linewidth=2,   # Line width 
         
         # Marker parameters 
         marker= 'o',             # Marker for data points
         markersize= 8,           # Marker size
         markerfacecolor= 'blue', # Marker fill color
         markeredgecolor= 'black',# Marker edge color
         markeredgewidth= 1       # Marker edge thickness    
        )

# Adding titles and labels
plt.title("Square Numbers Line Graph")
plt.xlabel("X values")
plt.ylabel("Y = X^2")

# Grid and legend
plt.grid(True)          # Show grid
plt.legend(["y = x^2"]) # Add legend

# Show the plot
plt.show()
```

### We can also plot multiple lines on a single graph

```python

# X values (shared)
x = [0, 1, 2, 3, 4, 5]

# Different Y values
y1 = [0, 1, 4, 9, 16, 25]     # y = x^2
y2 = [0, 1, 2, 3, 4, 5]       # y = x
y3 = [0, 1, 8, 27, 64, 125]   # y = x^3

# Plot each line with custom styles
plt.plot(x, y1, label="y = x^2", color='green', linestyle='--', marker='o')
plt.plot(x, y2, label="y = x", color='blue', linestyle='-', marker='s')
plt.plot(x, y3, label="y = x^3", color='red', linestyle='-.', marker='^')

# Add labels and title
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.title("Multiple Line Graphs")

# Show legend and grid
plt.legend()
plt.grid(True)

# Display the graph
plt.show()
```

### what if the x-values are not common to each line

```python
import matplotlib.pyplot as plt

# Line 1: Quadratic, with its own x values
x1 = [0, 1, 2, 3, 4]
y1 = [200, 100, 150, 290, 300]

# Line 2: Linear, different x range
x2 = [2, 3, 4, 5, 6]
y2 = [250, 200, 100, 80, 50]

# Line 3: Cubic, different and non-aligned x values
x3 = [1, 3, 5, 7]
y3 = [1, 27, 125, 343]

# Plot each line separately with its own x and y
plt.plot(x1, y1, label="Forella", color='green', marker='o')
plt.plot(x2, y2, label="Coops", color='blue', marker='s')
plt.plot(x3, y3, label="Cambium", color='red', marker='^')

# Labels and styling
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.title("Multiple Lines with Independent X Points")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Example data
clubs = ['Science', 'Drama', 'Art', 'Music', 'Sports']
students = [45, 30, 25, 40, 50]

# Create bar graph
plt.bar(clubs, students,
        color='skyblue', 
        edgecolor='black', 
        width=0.7)

plt.xlabel('Clubs')
plt.ylabel('Number of Students')
plt.title('Number of Students in School Clubs')
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```
